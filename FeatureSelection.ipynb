```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
import pandas as pd
import matplotlib.pyplot as py
```

```python
dataset  = pd.read_csv('train.csv')
```

```python
X= dataset.iloc[:,0:20]
y= dataset.iloc[:,-1]
```

```python
selectBest = SelectKBest(score_func=chi2)
y_fit = selectBest.fit(X,y)
```

```python
score = pd.DataFrame(y_fit.scores_)
column = pd.DataFrame(X.columns)
```

```python
concatScoreColumn = pd.concat([score,column], axis = 1)
```

```python
concatScoreColumn.columns=['score','column']
concatScoreColumn.nlargest(10, columns=['score'])
```
